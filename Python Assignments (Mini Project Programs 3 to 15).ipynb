```python
###3. SHOPPING CART SYSTEM

items = {
    "Apple": 50,
    "Banana": 20,
    "Milk": 40,
    "Bread": 35
}

cart = []

while True:
    print("\n1.View Items\n2.Add Item\n3.Remove Item\n4.View Cart\n5.Total Bill\n6.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        for item, price in items.items():
            print(item, price)

    elif choice == "2":
        item = input("Enter item name: ")
        if item in items:
            cart.append(item)

    elif choice == "3":
        item = input("Enter item to remove: ")
        if item in cart:
            cart.remove(item)

    elif choice == "4":
        print(cart)

    elif choice == "5":
        total = sum(items[i] for i in cart)
        if total > 100:
            total *= 0.9
        print("Total:", total)

    elif choice == "6":
        break
```

```python
###4. CONTACT BOOK

contacts = {}

def add():
    name = input("Name:")
    phone = input("Phone:")
    email = input("Email:")
    contacts[name] = {"phone": phone, "email": email}

def display():
    for name in sorted(contacts):
        print(name, contacts[name])

while True:
    print("1.Add 2.Display 3.Exit")
    ch = input()

    if ch == "1":
        add()
    elif ch == "2":
        display()
    else:
        break
```

```python
###5. LIBRARY MANAGEMENT

books = {"Python": True, "Java": True}

while True:
    print("1.Add 2.Issue 3.Return 4.View 5.Exit")
    ch = input()

    if ch == "1":
        name = input("Book:")
        books[name] = True

    elif ch == "2":
        name = input("Book:")
        if books.get(name):
            books[name] = False

    elif ch == "3":
        name = input("Book:")
        books[name] = True

    elif ch == "4":
        print(books)
    else:
        break
```

```python
###6. BANKING LEDGER

transactions = []
balance = 0

def add():
    global balance
    t = input("Type:")
    amt = int(input("Amount:"))

    transactions.append({"type": t, "amount": amt})

    if t == "credit":
        balance += amt
    else:
        balance -= amt

while True:
    print("1.Add 2.View")
    ch = input()

    if ch == "1":
        add()
    else:
        print(transactions)
        print(balance)
    else:
    break
```

```python
####7. INVENTORY SYSTEM

inventory = {}

name = input("Product:")
qty = int(input("Quantity:"))
price = int(input("Price:"))

inventory[name] = {"qty": qty, "price": price}

total = sum(v["qty"] * v["price"] for v in inventory.values())
print(total)
```

```python
#####8. QUIZ GAME

quiz = {
    "Capital of India": "Delhi",
    "2+2": "4"
}

score = 0

for q, a in quiz.items():
    ans = input(q)
    if ans == a:
        score += 1

print(score)
```
